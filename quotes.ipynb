```python
import requests
from lxml import html
all_data = []
for i in range(1,11):
    response = requests.get(f"https://quotes.toscrape.com/page/{i}/")
    tree = html.fromstring(response.content)
    quotes = tree.xpath("//div[@class='quote']")
    for q in quotes:
        data = {}
        data['Quotes'] = q.xpath(".//span[@class='text']")[0].text
        data['Author'] = q.xpath(".//small[@class='author']")[0].text
        data['Link'] = q.xpath(".//span/a")[0].get("href")
        print(data)
        all_data.append(data)
```

```python
import pandas as pd
pd.DataFrame(all_data).to_csv("quotes.csv")
```
